```python
import gym
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
env = gym.make('simworld:gridworld-jump-v0')
```

### Policy evaluation and improvement using dynamic programming

Using uniform policy, on one run of policy evaluation


Reaching the optimal policy using bellman optimal equation that is using greedy policy improvement.

We will try to reach this goal 



```python
class agent():
    def __init__(self, env):
        self.env = env
        self.V = np.zeros(len(env.getStates()))
        self.actions = env.getActions()
        self.policy = (np.ones((len(self.V), len(self.actions))))/len(self.actions)
        self.discount = 0.9
        self.threshold = 0.001
        
        
    def policy_evaluation_sweep(self):
        skip = []
        num = 0
        while True:
            delta = 0
            num += 1
            if num %100 == 0:
                print("episodes completed", num, "states completed", len(skip))
            for n, i in enumerate(self.V):
                if n in skip:
                    continue
                v = 0
                env.setState(n)
                for j in self.actions:
                    reward, next_state = env.step(j)
                    target = (reward + (self.discount*self.V[next_state]))
                    v += self.policy[n][j] * target
                delta = max(delta, abs(self.V[n] - v))
                self.V[n] = v
                
            if delta < self.threshold:
                break
        print("episodes completed", num)
        return self.V
    
    def action_selection(self, policy):
        return np.random.choice(a=self.actions,p=policy)
    
    def policy_improvment(self):
        policy_stable = True
    
        for n, i in enumerate(self.V):
            old_policy = self.policy[n].copy()
            new_policy = []
            env.setState(n)
            for j in self.actions:
                reward, next_state = env.step(j)
                target = (reward + (self.discount*self.V[next_state]))
                new_policy.append(target)
            
            new_p = np.zeros(len(self.actions))
            new_p[np.where(new_policy == np.max(new_policy))] = 1
            new_p = new_p/sum(new_p)

            self.policy[n] = new_p.copy()
            new_action = self.action_selection(self.policy[n])

            
            if (old_policy != self.policy[n]).any():
#                 print("policy change", old_policy, self.policy[n])
                policy_stable = False
        
        return policy_stable
        
                    
        
```

### Policy Iteration


```python
A = agent(env)
policy_stable = False
Vs = []
ps = []
while not policy_stable:
    ps.append(A.policy.copy())
    V = A.policy_evaluation_sweep()
    Vs.append(V.copy())
    policy_stable = A.policy_improvment()
    
```

```python
fig, axes = plt.subplots(len(Vs), 2, figsize=(25, 25))
# fig, axes = plt.subplots(2, 2, figsize=(25, 25))
for i in range(len(Vs)):
    sns.heatmap(Vs[i].reshape(5,5), ax=axes[i][0], annot=True, fmt=".1f", vmin=0, cmap="Blues", square=True, xticklabels=False, yticklabels=False)
    axes[i][0].set_title("iter = {}".format(i+1))
    sns.heatmap(ps[i].T, ax=axes[i][1], annot=True, fmt=".1f", vmin=1, vmax=0, cmap="Blues", square=True, yticklabels=False)
```

```python
class QAgent():
    def __init__(self, props):
        self.env = props['env']
        
        self.n_states = props['n_states']
        self.n_actions = props['n_actions']
        
        self.Q = np.zeros((self.n_states, self.n_actions))
        self.actions = props['actions']
        ##
        
        self.policy = props['policy']
        self.epsilon = props['epsilon']
        self.discount = props['discount']
        self.threshold = props['threshold']
        
        
    def policy_evaluation_sweep(self):
        skip = []
        num = 0
        while True:
            delta = 0
            num += 1
            if num %100 == 0:
                print("episodes completed", num, "states completed", len(skip))
            for n in range(self.n_states):
                for j in self.actions:
                    env.setState(n)
                    reward, next_state = env.step(j)
                    total_expected_reward = 0
                    for k in self.actions:         
                        total_expected_reward += (self.policy[next_state][k] * self.Q[next_state][k])
                    target = (reward + (self.discount*total_expected_reward))
                    q = self.policy[n][j] * target
                    delta = max(delta, abs(self.Q[n][j] - q))
                    self.Q[n][j] = q
                    
            if delta < self.threshold:
                break
        print("episodes completed", num)
        return self.Q
    
    def action_selection(self, policy):
        return np.random.choice(a=self.actions,p=policy)
    
    def policy_improvment(self):
        policy_stable = True
        for n in range(self.n_states):
            old_policy = self.policy[n].copy()
            new_policy = []
            for j in self.actions:
                env.setState(n)
                reward, next_state = env.step(j)
                total_expected_reward = 0
                for k in self.actions:
                    total_expected_reward += (self.policy[next_state][k] * self.Q[next_state][k])
                target = (reward + (self.discount*total_expected_reward))
                new_policy.append(target)
            
            new_p = np.zeros(len(self.actions))
            new_p += (self.epsilon/(len(self.Q[n])))
            new_p[np.where(new_policy == np.max(new_policy))] += (1 - self.epsilon)/(len(np.where(new_policy == np.max(new_policy))[0]))
            
            self.policy[n] = new_p.copy()
            new_action = self.action_selection(self.policy[n])
            
            if (old_policy != self.policy[n]).any():
                policy_stable = False
        
        return policy_stable
    
    
    def policy_greedy_evaluation_sweep(self):
        skip = []
        num = 0
        while True:
            delta = 0
            num += 1
            if num %100 == 0:
                print("episodes completed", num, "states completed", len(skip))
            for n in range(self.n_states):
                for j in self.actions:
                    env.setState(n)
                    reward, next_state = env.step(j)
                    
                    total_expected_reward = 0
                    max_action = -999999
                    for k in self.actions:
                        freturn = (self.policy[next_state][k] * self.Q[next_state][k])
                        if(max_action < freturn):
                            max_action = freturn
                        
                    target = reward + (self.discount*max_action)
                    q = self.policy[n][j] * target
                    delta = max(delta, abs(self.Q[n][j] - q))
                    self.Q[n][j] = q
                    
            if delta < self.threshold:
                break
        
        print("episodes completed", num)
        return self.Q
```

```python
#policy iteration
states = env.getStates()
actions = env.getActions()
n_states = len(states)
n_actions = len(actions)
policy = np.ones((n_states,n_actions))/n_actions

props = {
    'env': env,
    'states' : states,
    'actions': actions,
    'n_states': n_states,
    'n_actions': n_actions,
    'policy': policy,
    'discount': 0.9,
    'threshold': 0.001,
    'epsilon': 0.3
}

A = QAgent(props)
policy_stable = False
Qs = []
ps = []
while not policy_stable:
    
    ps.append(A.policy.copy())
    Q = A.policy_evaluation_sweep()
    Qs.append(Q.copy())
    policy_stable = A.policy_improvment()

    
fig, axes = plt.subplots(len(Qs), 2, figsize=(25, 25))
# fig, axes = plt.subplots(2, 2, figsize=(25, 25))

for i in range(len(Qs)):
    sns.heatmap(Q.T, ax=axes[i][0], annot=True, fmt=".1f", vmin=0, cmap="Blues", square=True, xticklabels=False, yticklabels=False)
    axes[i][0].set_title("iter = {}".format(i+1))
    sns.heatmap(ps[i].T, ax=axes[i][1], annot=True, fmt=".1f", vmin=1, vmax=0, cmap="Blues", square=True, yticklabels=False)
```

```python
#value iteration
states = env.getStates()
actions = env.getActions()
n_states = len(states)
n_actions = len(actions)
policy = np.ones((n_states,n_actions))/n_actions

props = {
    'env': env,
    'states' : states,
    'actions': actions,
    'n_states': n_states,
    'n_actions': n_actions,
    'policy': policy,
    'discount': 0.9,
    'threshold': 0.001,
    'epsilon': 0.3
}

A = QAgent(props)
policy_stable = False
Qs = []
ps = []
while not policy_stable:
    
    ps.append(A.policy.copy())
    ##value iteration
    Q = A.policy_greedy_evaluation_sweep()
    Qs.append(Q.copy())
    policy_stable = A.policy_improvment()

    
fig, axes = plt.subplots(len(Qs), 2, figsize=(25, 25))
# fig, axes = plt.subplots(2, 2, figsize=(25, 25))

for i in range(len(Qs)):
    sns.heatmap(Q.T, ax=axes[i][0], annot=True, fmt=".1f", vmin=0, cmap="Blues", square=True, xticklabels=False, yticklabels=['up','right', 'down', 'left'])
    axes[i][0].set_title("iter = {}".format(i+1))
    sns.heatmap(ps[i].T, ax=axes[i][1], annot=True, fmt=".1f", vmin=1, vmax=0, cmap="Blues", square=True, yticklabels=['up','right', 'down', 'left'])
```
